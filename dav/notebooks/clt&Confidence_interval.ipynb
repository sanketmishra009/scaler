```python
import pandas as pd
df = pd.read_csv('../data/weight-height.csv' )
df.columns = ['gender' , 'height' , 'weight']
# df = df.iloc[:,1:]
df
```

```python
means = []
fig , axes = plt.subplots(2,2, figsize = (12,12))
# print(axes)
for ind , ss in enumerate([5 , 10 , 50 ,100 ]):
    # print(ind / 2 , ind % 2)
    axis = axes[int(ind / 2),ind % 2]
    for i in range(200):
        mean = np.mean(df.sample(ss)['height'])
        means.append(mean)
    sns.histplot(means , bins = 30 , kde = True , ax = axis)
    axis.set_title('sample size:'+str(ss))
```

```python
#plot dist of height and weight by gender
fig , axes = plt.subplots(1,2 , figsize = (10,6))
# print(axes)
sns.histplot(df , x = 'height' , hue = 'gender' , kde = True , bins = 30 , ax = axes[0])
axes[0].set_title('Distribution of heights by gener')
axes[0].set_xlabel('Height (cm)')
axes[0].set_ylabel('Frequency')


#plot dist of weight by gender
sns.histplot(df , x = 'weight' , hue = 'gender' , kde = True , bins = 30 , ax = axes[1])
axes[1].set_title('Distribution of weights by gener')
axes[1].set_xlabel('weight (kg)')
axes[1].set_ylabel('Frequency')
plt.show()
```
