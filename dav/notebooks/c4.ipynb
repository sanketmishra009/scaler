```python
import pandas as pd
import numpy as np

np.set_printoptions(suppress = True , precision = 2)
```

```python
orders = pd.read_csv('../data/orders.csv')
products = pd.read_csv('../data/products.csv')
orders.head()
```

```python
print(f'{orders.shape=} , {products.shape=}')
```

```python
products.head()
```

```python
df = pd.merge(orders , products , on='product_id' , how='left')
df.head()
```

```python
#extract the price:
def fn(val):
    if(pd.isnull(val)):return val
    return float(str(val[1:]).replace(',' , ''))

df['actual_price'] = df['actual_price'].apply(fn)
df['discounted_price'] = df['discounted_price'].apply(fn)
df
```

```python
df['actual_price'].unique()
```

## calculate discount%:


```python
# df['actual_price'] , df['discounted_price']
def fn(row):
    return ((row['actual_price'] - row['discounted_price']) / row['actual_price']) * 100

df['calculated_discount'] = df.apply(fn , axis = 1)
df.head()
```

```python
df.groupby('category')['calculated_discount'].mean()
```

```python
def fn(col):
    return col.mean()

def fnn(col):
    # print(col)
    return col.mean()
df.groupby('category').agg({"rating":[fn , fnn , 'min' , 'max'] , "calculated_discount":[fn , fnn]})
```

```python
#apply a fn to the entire group:
def fgroup(group):
    return group['actual_price'].mean() - group['discounted_price'].mean()
df.groupby('category').apply(fgroup)
```

### IDIOMS in pandas.


```python
orders.head()
```

```python
import timeit
```

```python
def idiomatic():
    df = orders
    return (df.where(df['user_name'].str.startswith('M'))
    .dropna()
    .set_index('product_id')
    .rename(columns={'user_name':'userNmae'}))

timeit.timeit(idiomatic , number = 10)
```

```python
def nonidiomatic():
    df = orders
    df = df[df['user_name'].str.startswith('M')]
    df.set_index('product_id' , inplace = True)
    df.rename(columns = {'user_name':'userName'})
    return df
timeit.timeit(nonidiomatic , number = 10)
```

## Group by


```python
df = pd.merge(orders , products , on = 'product_id' , how = 'left')
df.head()
```

```python
for group , frame in df.groupby('category'):
    print('group: ', group , '\tmax_rating: ', frame['rating'].max())
```

```python
df.groupby('category')['rating'].value_counts().sort_values()
```

## groupby according to index and passing a function to the index as well.


```python
df = df.set_index(['category' , 'rating'])
df.head()
```

```python
for group , frame in df.groupby(level = (0,1)):
    print(group)
```

```python
def fn(item):
    if(item[1] > 4):
        return (item[0] , 1)
    return (item[0] , 0)
for group , frame in df.groupby(by = fn):
    print(group)
```
