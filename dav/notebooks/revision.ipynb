```python
def gettime(timee):
    hour = timee.hour
    if(hour < 6):return 'morning'
    elif(hour < 12): return 'day'
    elif(hour < 18): return 'afternoon'
    else: return 'night'

df['timeOday'] = pd.to_datetime(df['pickup']).apply(gettime)
df
```

```python
df.groupby('timeOday')['pickup'].count().reindex(['morning', 'day', 'afternoon', 'night'] , fill_value=0).sort_values(ascending = False)
```
