```python
# QQ plot
import numpy as np
import scipy.stats as stats
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
snormal = np.random.normal(loc = 0 , scale = 1 , size = 10000)
nquants = []
for i in range(0 , 101):
    nquants.append(np.percentile(snormal , i))
nquants
```

```python
rnormal = np.random.normal(loc = 20 , scale = 50 , size = 10000)
rquants = []
for i in range(0 , 101):
    rquants.append(np.percentile(rnormal , i))

rquants
```

```python
sns.scatterplot(x = nquants , y = rquants)
plt.plot(np.arange(-4, 4) , np.arange(-4, 4) * 50 + 20 , color = 'red')
plt.title('Q-Q plot')
plt.show()
```

```python
runiform = np.random.uniform(low = -1 , high = 1 , size = 10000)
ruquants = []
for i in range(0 , 101):
    ruquants.append(np.percentile(runiform , i))
```

```python
sns.scatterplot(x = nquants , y = ruquants)
plt.plot(np.arange(-4 , 5) , np.arange(-4 , 5) , color = 'red')
plt.title('Q-Q plot')
plt.show()
```
