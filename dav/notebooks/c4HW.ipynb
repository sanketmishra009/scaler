## Given two dataframes customer (containing details of customer) and orders (having details of orders).

Also, name is given which is a customer's name.

Perform the following operation:

    Merge the dataframe such that the resultant dataframe should contain records of all customer ids present in customer dataframe.

    Calculate the total order amount for given customer name

Return the merged dataframe and the sum amount.


```python
data = {
    'cust_id': [101, 102, 103, 104],
    'name': ['rick', 'morty', 'pickle', 'jerry']
}

customers = pd.DataFrame(data)
data = {
    'order_id': ['OR1', 'OR3', 'OR23', 'OR42'],
    'cust_id': [102, 105, 101, 102],
    'amount': [1200, 650, 120, 989]
}

orders = pd.DataFrame(data)
customers.head() , orders.head()
```

```python
mdf = pd.merge(customers , orders , how='left' , on='cust_id')
mdf
```

```python
mdf.groupby('name')['amount'].agg('sum').loc['morty']
```

### Create a new column time_of_day containing the labels for pickup time based on the following rules-

[0 am to 6 am) - morning

[6 am to 12 pm) - day

[12 pm to 6 pm) - afternoon

[6 pm to 0 am) - night

For all the intervals above, left limit is included but right limit is not included in the range.

Return a pandas series containing the number of pickups for each time_of_day. Sort the counts in descending order.


```python
df = pd.DataFrame([['2019-03-23 20:21:09', '2019-03-23 20:27:24', 1, 1.6], ['2019-03-04 16:11:55', '2019-03-04 16:19:00', 1, 0.79], ['2019-03-27 17:53:01', '2019-03-27 18:00:25', 1, 1.37], ['2019-03-10 01:23:59', '2019-03-10 01:49:51', 1, 7.70], ['2019-03-30 13:27:42', '2019-03-30 13:37:14', 3, 2.16]], columns = ['pickup', 'dropoff', 'passengers', 'distance'])
df
```

```python
# times = pd.to_datetime(df['pickup'])
def gettime(timee):
    hour = timee.hour
    if(hour < 6):return 'morning'
    elif(hour < 12): return 'day'
    elif(hour < 18): return 'afternoon'
    else: return 'night'
df['time_of_day'] = pd.to_datetime(df['pickup']).apply(gettime)
df
```

```python
df.pivot(index = 'pickup' , columns = 'time_of_day')
```

```python
all_categories = ['morning', 'day', 'afternoon', 'night']
df.groupby('time_of_day')['pickup'].count().reindex(all_categories , fill_value=0).sort_values(ascending=False)
```

### Q4. Problem Description:

You are provided with a DataFrame sales_data containing information about sales transactions. Each row represents a single transaction with the following columns:

    Date: The date of the transaction.
    Region: The region where the transaction occurred.
    Product: The product that was sold.
    Sales: The amount of sales for that transaction.

Your task is to create a pivot table from sales_data that shows the total sales amount each region. You need to write a function generate_pivot_table(data) that takes the DataFrame sales_data as input and returns the pivot table.

Your function should:

    Create a pivot table with 'Region' as the index and the sum of 'Sales' as the values.
    Handle missing values, filling them with zeros.
    Sort the resulting pivot table rows and columns alphabetically.

Your function should return the pivot table as a DataFrame.

    Refer to sample input and output for better understanding.


```python
data = {
    'Date': ['2024-01-01', '2024-01-01', '2024-01-02', '2024-01-02'],
    'Region': ['East', 'West', 'East', 'West'],
    'Product': ['A', 'B', 'A', 'B'],
    'Sales': [100, 200, 150, 250]
}

df = pd.DataFrame(data)
df
```

```python
df.pivot_table(index = 'Region' , values = 'Sales' , aggfunc='sum' , dropna=False , fill_value=0)
```

```python
df.groupby(by='Region').agg({'Sales':'sum'})
```

You are given a dataframe df containing life expectancies of different countries for different years.

Melt all the year columns into a single column. Set variable name as 'Year' and value name as 'Life_Expectancy'.

Find 5 years with the maximum average life expectancy.

Return a dataframe containing 5 rows with Year as index and Life Expectancy as a column, sorted in descending order of average life expectancy.

Note that the Life Expectancy column in the output represents the avarage life expectencies for different years.


```python
df = pd.DataFrame([['Canada', 70.6, 78.3, 80.2, 82.3, 81.4, 80.3], ['France', 71.8, 74.7, 79.32, 77.2, 80.0, 81.2], ['Germany', 74.8, 74.7, 77.32, 77.2, 80.0, 81.2], ['Japan', 74.9, 74.7, 77.32, 77.8, 82.0, 81.2], ['Korea', 74.1, 74.7, 77.32, 77.2, 80.0, 83.2]], columns = ['Country', 2000, 2001, 2002, 2003, 2004, 2005])

df
```

```python
dfm = df.melt(id_vars='Country' , var_name='Year' , value_name='Life_Expectancy')
dfm
```

```python
dfm.groupby('Year').agg({'Life_Expectancy' : 'mean'}).sort_values(by= 'Life_Expectancy' , ascending=False)
```
