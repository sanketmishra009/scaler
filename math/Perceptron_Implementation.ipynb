```python
def plot_line(W,w0,x,y):
  min = x[:,0].min()
  max = x[:,0].max()

  w1 = W[0]#1
  w2 = W[1]#1
  w0 = w0

  slope = -w1/w2
  intercept = -w0/w2

  x_values = np.linspace(min,max,100)

  y_values = slope*x_values+intercept


  sns.scatterplot(x = x[:,0], y = x[:,1], hue = y)
  plt.plot(x_values,y_values,c='r')
  plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.datasets import load_iris
from sklearn.preprocessing import StandardScaler
from sklearn.model_selection import train_test_split

# Load iris dataset and take only class 0 and 1
iris = load_iris()
X = iris.data
y = iris.target

# Keep only class 0 and 1
X = X[y != 2]
y = y[y != 2]

# Use only two features for 2D plot
X = X[:, :2]  # sepal length and sepal width

# Standardize features
scaler = StandardScaler()
X = scaler.fit_transform(X)

# Convert labels from (0,1) to (-1,+1) for perceptron
y = np.where(y == 0, -1, 1)

# Initialize weights and bias
W = np.zeros(X.shape[1])
w0 = 0
epochs = 20

# Perceptron training
def train_perceptron(X, y, W, w0, epochs):
    for epoch in range(epochs):
        misclassified = 0
        for xi, yi in zip(X, y):
            if np.sign(np.dot(W, xi) + w0) != yi:
                W += yi * xi
                w0 += yi
                misclassified += 1

        print(f"Epoch {epoch+1} - Misclassified: {misclassified}")
        plot_decision_boundary(W, w0, X, y, epoch+1)
        if misclassified == 0:
            break
    return W, w0

# Plotting decision boundary
def plot_decision_boundary(W, w0, X, y, epoch):
    plt.figure(figsize=(6, 4))
    for xi, yi in zip(X, y):
        if yi == -1:
            plt.scatter(xi[0], xi[1], color='red', marker='o', label='Setosa' if 'Setosa' not in plt.gca().get_legend_handles_labels()[1] else "")
        else:
            plt.scatter(xi[0], xi[1], color='blue', marker='x', label='Versicolor' if 'Versicolor' not in plt.gca().get_legend_handles_labels()[1] else "")

    # Decision boundary: W1*x1 + W2*x2 + w0 = 0 => x2 = -(W1*x1 + w0)/W2
    x_vals = np.linspace(X[:,0].min()-1, X[:,0].max()+1, 100)
    y_vals = -(W[0]*x_vals + w0)/W[1]
    plt.plot(x_vals, y_vals, 'k--', label='Decision Boundary')

    plt.xlabel("Sepal Length (standardized)")
    plt.ylabel("Sepal Width (standardized)")
    plt.title(f"Perceptron - Epoch {epoch}")
    plt.legend()
    plt.grid(True)
    plt.show()

# Train and visualize
final_W, final_w0 = train_perceptron(X, y, W, w0, epochs)
```
